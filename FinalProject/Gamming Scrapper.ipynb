```python
# To obtain all games released on the specified platforms (Nintendo Switch, PlayStation 5, PlayStation 4, Xbox Series X/S, Xbox One) from 2010 to 2020, and store the data in a DataFrame and then save it to a CSV file, you can use the code below:

import pandas as pd
import requests
import datetime

# Replace 'YOUR_API_KEY' with your actual RAWG API key
API_KEY = API
BASE_URL = 'https://api.rawg.io/api/'

def get_games_for_platforms(platform_ids, start_year=2022, end_year=2022):
    search_url = f"{BASE_URL}games"
    start_date = f"{start_year}-01-01"
    #     end_date = f"{end_year}-12-31" #will read 2 whole years in 2000-2001 as 24 months
    end_date = f"{start_year}-12-31" #will read 2 whole years in 2000-2001 as 24 months
    #     end_date = f"{end_year}-01-01" #12 months exactly
    if(end_year>=2024):
        end_date = f"2023-07-30"   # 7 months of 2023
        start_date = f"2023-01-01"


    params = {
        'platforms': ','.join(str(platform_id) for platform_id in platform_ids),
        'key': API_KEY,
        'dates': f"{start_date},{end_date}",
        'page_size': 200,
        'page': 1
    }

    all_games = []

    while True:
        response = requests.get(search_url, params=params)

        if response.status_code == 200:
            data = response.json()
            games_on_page = data['results']
            all_games.extend(games_on_page)
            #Save a checkpoint
            df = pd.DataFrame(all_games)
            df.to_csv(f"chkpt_games_{start_year}_to_{end_year}_3.csv" , index=False) #,mode='a', header=False,

            if data['next']:
                params['page'] += 1
            else:
                break
        else:
            print(f"Failed to fetch game data. Error code: {response.status_code}")
            print(response.reason)
            print(response)
            print()
            break

    return all_games

if __name__ == "__main__":
    # Platform IDs: Nintendo Switch, PlayStation 5(187), PlayStation 4(18), Xbox Series X/S(186), Xbox One (1), PC(4)
#     platform_ids = [7, 187, 186, 18, 1, 4]
    platforms_dic_s={ 'Nintendo Switch':7, 'PlayStation 5':187, "PlayStation 4":18, "Xbox Series X/S":186, "Xbox One":1, "PC":4}
    platforms_dic={7: 'Nintendo Switch',187: 'PlayStation 5',18: 'PlayStation 4',186: 'Xbox Series X/S',1: 'Xbox One',4: 'PC'}
    platform_ids = [7, 187, 186, 18, 1, 4]
    START_YEAR = 2018
    END_YEAR = 2023
    
    
    for i in range(START_YEAR, END_YEAR):
        start_year = i
        end_year = i+1
        games_for_platforms = get_games_for_platforms(platform_ids, start_year, end_year)
        print("===========",start_year,end_year,"============")
        
        if games_for_platforms:
            print(f"Total games found for the specified platforms ({([v for k,v in platforms_dic.items() if k in platform_ids])}) and years({start_year}-{end_year}): {len(games_for_platforms)}")

            # Convert the list of games to a DataFrame
            df = pd.DataFrame(games_for_platforms)

            # Save the DataFrame to a CSV file
            df.to_csv(f"games_{start_year}_to_{end_year}.csv", index=False)

            print(f"Data saved to 'games_{start_year}_to_{end_year}.csv'\n")
        else:
            print("No games found for the specified platforms and years.\n")
        print("=================================\n\n")
            

# This code will fetch all games for the specified platforms released between 2010 and 2020 using the RAWG API, store the data in a pandas DataFrame, and then save the DataFrame to a CSV file named 'games_2010_to_2020.csv' in the same directory as the script. The CSV file will contain information about each game, including its name, release date, platform details, etc.
# Failed to fetch game data. Error code: 404
# Total games found for the specified platforms and years: 10000
# Data saved to 'games_2010_to_2020.csv'
```

```python
import requests

# Replace 'YOUR_API_KEY' with your actual RAWG API key
# API_KEY = 'YOUR_API_KEY'
BASE_URL = 'https://api.rawg.io/api/'

def get_platforms_list():
    platforms_url = f"{BASE_URL}platforms"
    params = {
        'key': API_KEY
    }

    response = requests.get(platforms_url, params=params)

    if response.status_code == 200:
        data = response.json()
        platforms = data['results']
        return platforms
    else:
        print(f"Failed to fetch platforms data. Error code: {response.status_code}")
        return []

if __name__ == "__main__":
    platforms_list = get_platforms_list()

    if platforms_list:
        print("List of platforms:")
        for platform in platforms_list:
            print(f"Platform ID: {platform['id']}, Platform Name: {platform['name']}")
        df_plats = pd.DataFrame(platforms_list)
    else:
        print("No platforms found.")
```
