
## Python Lambda
Learn some lambda  
Is an anonymoys small anonymous funciton  
A lamba funciton can take any number of argument , but can only have one expression,   
the expresion executed and the result is returned  



```python
# IE, add 15 to argument a and return the result
x = lambda a : 15 + a
print(x(10))
```

```python
# Multiply a,b,c, arguments and t return the result
x = lambda a,b,c : a * b *  c
print(x(2,4,5))
```

```python
# Use two lambdas connected
import math
 #input multiply by 2
l1 = lambda a : a*2   
# sqrt of a
l2 =  lambda a : math.sqrt(l1(a))

print(l2(12.5))
```

### Finding min max in matrix

```python
# using max and min 
matrix = np.array([[1,2,30],[2,4,5],[7,8,9]])

print("Max:",np.max(matrix))
print("Min:",np.min(matrix))

```

```python
print(matrix)
# 0 meands row wise
# Use the axis paramenter to apply the operation along a certain acis

print("Max on Row:", np.max(matrix, axis=0))
print("Max on Columns:", np.max(matrix, axis=1))


```

```python
# To get just the max value on the # columns
for i in range(0,len(matrix[0])):
    print(f"Max on Column index {i}:", np.max(matrix[:,i]))
```

## Calculate statistical values


```python
# Using mean, var and std methods on numpy
print("np.mean(matrix):\n\t",np.mean(matrix))

# Variance
print("np.var(matrix):\n\t",np.var(matrix))

# Standard Deviation
print("np.std(matrix):\n\t",np.std(matrix))

#using the axis paramenter to apply the operation along a certain axis
print("np.mean(matrix, axis =0):\n\t",np.mean(matrix, axis=0))

# get strd in column wise
print("np.std(matrix, axis=1)",np.std(matrix, axis=1))
```
